```python
import numpy as np
import matplotlib.pyplot as plt

# Fixar semente para reprodutibilidade
np.random.seed(42)

# Número de dias simulados
n_dias = 250  # aproximadamente 1 ano de pregões

# Simulação de retornos diários (%)
btc = np.random.normal(loc=0.15, scale=2.0, size=n_dias)   # Bitcoin: menor volatilidade
eth = np.random.normal(loc=0.20, scale=3.0, size=n_dias)   # Ethereum: volatilidade média
sol = np.random.normal(loc=0.30, scale=5.0, size=n_dias)   # Altcoin: maior volatilidade

# Organizar os dados
dados = [btc, eth, sol]
labels = ['Bitcoin', 'Ethereum', 'Solana']

# Criar o boxplot
plt.figure(figsize=(8, 5))
plt.boxplot(dados, labels=labels, showfliers=True)

# Ajustes visuais
plt.title('Boxplot de Retornos Diários Simulados (%)')
plt.ylabel('Retorno Diário (%)')
plt.xlabel('Criptomoeda')
plt.grid(axis='y')

plt.show()
```
