```python
# carregamento de dados (json)
df_customers = pd.read_json('/content/customers_data.json')
df_products = pd.read_json('/content/products_data.json')
df_sales = pd.read_json('/content/sales_data.json')
```

```python
# pré-diagnóstico
df_customers.info() # info() mostra o nome das colunas - quantidade de dados por coluna - dataType
df_customers.head() # exibe os primeiros 5 dados
```

```python
df_products.info() # info() mostra o nome das colunas - quantidade de dados por coluna - dataType
df_products.head() # exibe os primeiros 5 dados
```

```python
df_sales.info() # info() mostra o nome das colunas - quantidade de dados por coluna - dataType
df_sales.head() # exibe os primeiros 5 dados
```

```python
# Limpar os dados - Costumers
media_idade = df_customers['age'].mean() # calcula a média
# para substituir um campo devemos:
# 1 - informar qual o campo
# 2 - falar que ele vai receber ele mesmo com a condição
df_customers['age']= df_customers['age'].fillna(media_idade)

#data invalida
df_customers['signup_date'] = pd.to_datetime(df_customers['signup_date'], errors= "coerce")

# padronizar genero
df_customers['gender'] = df_customers['gender'].replace("",pd.NA).str.upper()
#.replace("o que era", "o que vai ser").str.upper() para deixar tudo maiusculo
display(df_customers)
```

```python
# Limpar os dados - Products
# vamos trocar o basecost vazio pela media dos basecosts
media_basecost = df_products['base_cost'].mean()
df_products['base_cost'] = df_products['base_cost'].fillna(media_basecost)
display(df_products)
```

```python
# Limpar os dados - Sales
# o sales, vendas esta incorreto
df_sales['sales'] = pd.to_numeric(df_sales['sales'], errors= "coerce")

# vendas < 0
# localizar tudo que é menor que zero
# 1 - chama a base de dados com o atributo .loc
# 2 - o campo que eu quero
# 3 - informa a condição e o campo
# 4 - o que vai substituir
df_sales.loc[df_sales['sales']<=0, 'sales'] = pd.NA

# converter as datas, que estao com diferentes padronizacoes

df_sales ['date'] = pd.to_datetime(df_sales['date'], errors= "coerce")
print('quantidade de dados truncados')
print(df_sales['date'].isna().sum())

# arrumando a regiao
# para alterar um valor vazio que pode vir de diferentes maneiras (none"")
# podemos usar o comando .str.strip().str.title() para escrever o titulo que esta la
# se era vazio quer dizer que nao tem titulo, ou seja, mostra o texto NONE
df_sales['region'] = df_sales['region'].str.strip().str.title()

# arrumar o customer_type

df_sales['customer_type'] = df_sales['customer_type'].replace("",pd.NA)
display(df_sales)
```

```python
# ATENCAO - VOCE VAI EXCLUIR DADOS
# remover dados
df_sales_limpos = df_sales.dropna(subset=['sales','date'])
# o comando dropna(subset=["CAMPOS A SEREM DETELATADOS"])

# TABELAS DE CORRELACOES - TABELA DINAMICA DO EXCEL
# criar um campo de correlacao entre os consumidores
mapeamento_clientes = {
    "Novo": "C001",
    "Recorrente":"C002",
    "Corporativo":"C003"
}
#vamos colocar o novo campo
df_sales_limpos['customer_id'] = df_sales_limpos['customer_type'].map(mapeamento_clientes)
```

```python
# para ver as colunas
df_sales_limpos.columns
```

```python
df_consolidado = df_sales_limpos.merge(df_customers, on = 'customer_id')
```

```python
display(df_consolidado)
```

```python
df_consolidado = df_consolidado.rename(columns={"region_x":"region_sales"})
df_consolidado = df_consolidado.rename(columns={"region_y":"region_customer"})

# ATENCAO VOCE PODERIA EXCLUIR UMA DAS REGIOES
#df_consolidado = df_consolidado.dropna(columns=['region_y'])
display(df_consolidado)
```

```python
#quantidade de vendas por regiao

pd.pivot_table(
    # 1 - qual base de dados
    df_consolidado,
    # 2 - Quais valores
    values="quantity",
    # 3 - quais os index (legendas)
    index = 'region_sales',
    # função de agregação
    aggfunc="sum"
)
```

```python
# Ticket médio por Região
# correlacione o valor de venda por regiao
pd.pivot_table(
    df_consolidado,
    values=['sales', 'cost'],
    index = "region_sales",
    aggfunc="mean"
)
```
