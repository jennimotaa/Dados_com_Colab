```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# gerar um grafico de barras

```python
# diagrama de dispersao
# dados do data frame
x= [10,20,30,40,50]
y= [15,25,35,45,55]

#criar a visualização
plt.scatter(x,y, color = 'blue', label = "Vendas no tempo")
plt.title("Vendas no tempo", fontsize =14)
plt.xlabel("Tempo", fontsize=11)
plt.ylabel("Vendas",fontsize = 11)
plt.legend()
plt.grid(True)
#exibir a visualização
plt.show()
#
```

```python
# gerar um grafico de barras
# carregando os dados
categoria = ['A','B','C','D','E'] #STRING
valores = ['10','40','30','65','8'] # int
plt.bar(categoria,valores, color = 'black', edgecolor = "yellow")
plt.title('Grafico de Barras por categorias', fontsize = 14)
plt.xlabel('Categorias', fontsize=11)
plt.ylabel('Valores',fontsize = 11)
plt.grid(axis='y', linestyle = '--', alpha = 0.7)
```

```python
# grafico de linhas
x= [10,20,50,60,70]
y= [15,65,35,45,85]
#criar a visualização
plt.plot(x,y, color = 'blue',label = 'vendas mensais', linestyle= '-')
plt.title("Acompanhamento das vendas", fontsize =14)
plt.xlabel("Tempo", fontsize=11)
plt.ylabel("Vendas",fontsize = 11)
#plt.legend()
plt.grid(True)
#exibir a visualização
plt.show()
#
```

```python
#montar bloxplot
vendas_mensais = [300, 380, 360, 400, 450, 450, 500]

#plt.figure a gente pode usar pra box plot e também gráfico de pizza
#para gráfico de setores usa "em baixo" plt.pie
#para boxplot usa o plt.boxplot

plt.figure(figsize = (8, 6))
plt.boxplot(vendas_mensais, patch_artist=True,
            boxprops=dict(facecolor='skyblue', color='black'), medianprops=dict(color='red'))
```
