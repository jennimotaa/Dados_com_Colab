```python
# ETL - Extrair, transformar e carregar

import pandas as pd
import matplotlib.pyplot as plt
import json

df_producao = pd.read_json('/content/producao.json')

# maneira de exibir dado composto
# sempre que os dados nao podem ser exibidos em formato de tabela
# uso a funcao with open(caminho do arquivo) as arquivo

with open ('/content/producao.json') as arquivo:
  producao = json.load(arquivo)

df_producao = pd.DataFrame(producao['lotes'])
df_producao
```

```python
# regra de bernoulli (somente duas opcoes dar certo/errado)
# retirar uma amostra

# retirar um produto defeituoso dentre todos
# total / amostra
total_doces = df_producao ['doces'].sum()
total_defeitos = df_producao ['defeitos'].sum()
probabilidade = total_defeitos / total_doces
print(f'{probabilidade:.2%}')

# arredondamento:.0
# cientifico: .3
# financeiro: .8
# quanto a maior a quantidade de casas decimais, maior a precisão
```

```python
 # quantos defeitos esperamos em um lote?
qtd_lote = df_producao['doces'].mean()
print(qtd_lote)

df_qualidade = pd.read_json('/content/qualidade.json')
sensibilidade = df_qualidade['taxa_defeito']['sensibilidade']
print("o valor é:",sensibilidade)
```

```python
# gerar valores aleatórios
import numpy as np
# numpy é a biblioteca numerica do python
from numpy import random
# colocamos a biblioteca numpy e dela chamamos o modulo random

numero = random.randint(10) # ele faz do 0 a 9
print(numero +1)
# para numeros decimais
decimal = random.rand() # aleatorio entre 0 e 1
print(decimal)
```
